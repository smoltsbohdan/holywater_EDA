```python
import pandas as pd

# Отримання абсолютного шляху до файлу
file_path = f'C:/Users/smolt/PycharmProjects/holywater_EDA/data/installs_data.csv'
# Завантаження даних
df_of_installs = pd.read_csv(file_path)
```

```python
# Перевірка типів даних та кількість непорожніх значень в кожному стовпці
data_info = df_of_installs.info()
```

```python
# Перевірка унікальних значень в категоріальних стовпцях
unique_values = df_of_installs[['channel', 'medium', 'campaign']].apply(lambda x: x.nunique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Конвертація стовбця install_time в datetime
df_of_installs['install_time'] = pd.to_datetime(df_of_installs['install_time'])

# Виведення розподілу часу встановлення додатка
plt.figure(figsize=(12, 6))
sns.histplot(df_of_installs['install_time'], bins=30, kde=True)
plt.title('Розподіл часу встановлення додатка')
plt.xlabel('Дата та час')
plt.ylabel('Кількість встановлень')
plt.show()
```

```python
# Аналіз якості трафіку з різних каналів та кампаній
traffic_quality = df_of_installs.groupby(['channel', 'campaign']).size().unstack()
traffic_quality.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Кількість встановлень за каналами та кампаніями')
plt.xlabel('Канал')
plt.ylabel('Кількість встановлень')
plt.legend(title='Кампанія', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Розгляд ключових слів або контенту, що приводять до встановлення
keyword_analysis = df_of_installs.groupby('keyword').size().sort_values(ascending=False).head(10)
keyword_analysis.plot(kind='bar', figsize=(12, 6))
plt.title('Топ-10 ключових слів або контенту, що приводять до встановлення')
plt.xlabel('Ключове слово або контент')
plt.ylabel('Кількість встановлень')
plt.show()
```

```python
# Конвертація 'iap_item_price' в числовий тип
df_of_orders['iap_item_price'] = df_of_orders['iap_item_price'].replace(',', '', regex=True).astype(float)

# Конвертація 'discount_amount' в числовий тип
df_of_orders['discount_amount'] = df_of_orders['discount_amount'].replace(',', '', regex=True).astype(float)

# Середній чек
average_order_value = df_of_orders['iap_item_price'].mean()
# Сума знижок
discount_amount = df_of_orders['discount_amount'].sum()

print(f'Середній чек: {average_order_value}')
print(f'Загальна сума знижок: {discount_amount}')
```

```python
# Зміна типу даних стовпців 'event_time'
df_of_events['event_time'] = pd.to_numeric(df_of_events['event_time'], errors='coerce')
df_of_orders['event_time'] = pd.to_numeric(df_of_orders['event_time'], errors='coerce')

# З'єднання таблиць
merged_data = pd.merge(df_of_installs, df_of_costs, how='outer')
merged_data = pd.merge(merged_data, df_of_events, how='outer')
merged_data = pd.merge(merged_data, df_of_orders, how='outer')
```

```python
# Огляд об'єднаних даних
print(merged_data.head())
print(merged_data.info())
```

```python
# Вилучення ненумеричних стовпців
numeric_columns = merged_data.select_dtypes(include=['float64', 'int64']).columns
numeric_data = merged_data[numeric_columns]

# Додатковий аналіз об'єднаних даних
# Визначення, які змінні у цьому наборі даних взаємопов'язані між собою
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".1f", annot_kws={"size": 6})
plt.title('Кореляційна матриця')
plt.show()
```

```python
# Порівняння витрат та прибутку
costs_by_campaign = merged_data.groupby('campaign')['cost'].sum()
revenue_by_campaign = merged_data.groupby('campaign')['iap_item_price'].sum()

# Прибуток та витрати за кампанією
profit_by_campaign = revenue_by_campaign - costs_by_campaign
print(profit_by_campaign)
```

```python
# Визначення, що застосунок був встановлений на основі user_action_detail або event_type
merged_data['is_installed'] = (merged_data['user_action_detail'].notnull()) | (merged_data['event_type'].notnull())
```
